```python
import pandas as pd
df = pd.read_csv("weather_data2.csv")
df
```

```python
import numpy as np
df1 = df.replace(-99999, np.NaN)
df1
```

```python
df2 = df.replace({ #replace values in specific columns using dictionary
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0'
}, np.NaN)
df2
```

```python
df3 = df.replace({ #mapping dictionary
    -99999: 'blah',
    '0': 'blahblah'
})
df3
```

```python
df4 = df.replace('[A-Za-z]', 'L', regex=True) #replace using regex
df4
```

```python
df5 = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df5
```

```python
df6 = df5.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
df6
```
