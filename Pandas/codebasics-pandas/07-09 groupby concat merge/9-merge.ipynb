```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "singapore"],
    "temperature": [21,14,35, 31],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando", "taipei"],
    "humidity": [65,68,75,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city') #performs an inner join by default (intersection)
df3
```

```python
df4 = pd.merge(df1, df2, on="city", how="outer") #perform an outer join
df4
```

```python
df5 = pd.merge(df1, df2, how="left") #perform a left inner join
df5
```

```python
df6 = pd.merge(df1, df2, how="right") #right inner join
df6
```

```python
df7 = pd.merge(df1, df2, on="city", how="outer", indicator=True) #perform an outer join
df7
```
