```python
import pandas as pd
df = pd.read_csv("aapl_no_dates.csv")
df.head()
```

```python
rng = pd.date_range(start="6/1/2017", end="6/30/2017", freq='B') #B for business days
rng
```

```python
df.set_index(rng, inplace=True)
df
```

```python
%matplotlib inline
df.plot(y=["Open", "High", "Low", "Close"])
```

```python
df.Close.plot()
```

```python
df['Avg']=df[['Open','Close']].mean(axis=1)
df
```

```python
df.Avg.plot()
```

```python
df.asfreq('D', method='pad') #padding carries forward the values into empty rows
```
